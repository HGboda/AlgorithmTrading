```python
import numpy as np
import pylab as pl
from pykalman import UnscentedKalmanFilter
import warnings
bars = v20.barstmp_ret
global day 
day = 0
global day_idx1
day_idx1 = 0
global day_idx2
day_idx2 = 0

# initialize parameters
def transition_function(state, noise):
    
    try:
#         a = np.sin(state[0]) + state[1] * np.sin(noise[0])*1.0
        a = state[0]+np.sin(noise[0])*0.5
        b = state[1] + noise[1]
#         print a,b
    except Excpetion,e:
        print e
        PrintException()
    return np.array([a, b])

def observation_function(state, noise):
    try:
        C = np.array([[1, 0], [0, 1]])
    except Exception,e:
        print e
        PrintException()
    return np.dot(C, state) + noise

def PrintException():
    exc_type, exc_obj, tb = sys.exc_info()
    f = tb.tb_frame
    lineno = tb.tb_lineno
    filename = f.f_code.co_filename
    linecache.checkcache(filename)
    line = linecache.getline(filename, lineno, f.f_globals)
    print 'EXCEPTION IN ({}, LINE {} "{}"): {}'.format(filename, lineno, line.strip(), exc_obj)

    
def drawmain():
    # draw estimates
    with warnings.catch_warnings():
        warnings.filterwarnings('ignore', message='Polyfit may be poorly conditioned')
        warnings.filterwarnings('ignore', message='using a non-integer number instead of an integer will result in an error in the future')
        try:
            global day
            day = 0
            global day_idx1    
            day_idx= 0
            global day_idx2    
            day_idx2= 0
            
            transition_covariance = np.eye(2)
            random_state = np.random.RandomState(0)
            observation_covariance = [[0.1, 0], [0, 0.1]]
            initial_state_mean = [bars['bench'][1], 0]
            initial_state_covariance = [[0.1, 0], [0, 0.1]]
            
            # sample from model
            kf = UnscentedKalmanFilter(
                transition_function, observation_function,
                transition_covariance, observation_covariance,
                initial_state_mean, initial_state_covariance,
                random_state=random_state
            )
            xlist = []
            for bench in bars['bench'][1:]:
                xlist.append([float(bench),float(0.1)])
#             print np.array(xlist)
#             observations = np.array([bars['Close'].tolist(),0])
#             observations = xlist[:10]
#             print xlist[2]
#             observations = np.ma.array([xlist[0],xlist[1],xlist[2],xlist[2],xlist[2],xlist[2],xlist[2],xlist[2]])
            observations = np.ma.array(xlist)
#             states, observations = kf.sample(5, initial_state_mean)
#             print type(observations)
#             print observations
            # estimate state with filtering and smoothing
            filtered_state_estimates = kf.filter(observations)[0]
            smoothed_state_estimates = kf.smooth(observations)[0]
#             print filtered_state_estimates[:,0]
            pl.figure()
#             lines_true = pl.plot(states, color='b')
            lines_filt = pl.plot(filtered_state_estimates[:,0], color='r', ls='-')
            lines_smooth = pl.plot(smoothed_state_estimates[:,0], color='g', ls='-.')
            pl.legend((lines_filt[0], lines_smooth[0]),
                      ('filt', 'smooth'),
                      loc='lower left'
            )
            pl.show()
        except Exception,e:
            print e
            PrintException()
drawmain()
```

```python
def get_quantiles(pd, mu, sigma):

    discrim = -2 * sigma**2 * np.log(pd * sigma * np.sqrt(2 * np.pi))

    # no real roots
    if discrim < 0:
        return None

    # one root, where x == mu
    elif discrim == 0:
        return mu

    # two roots
    else:
        return mu - np.sqrt(discrim), mu + np.sqrt(discrim)
from scipy.stats import norm
x = bars['bench']
loc_in = x.mean()
scale_in = x.std()
print 'loc',loc_in,'scale',scale_in

pd = y = norm.pdf(x, loc=loc_in, scale=scale_in)    # for example
print get_quantiles(pd, loc_in, scale_in)
```
